```python
gnm = open('MN908947.txt', 'r')
```

```python
genome = []
for line in gnm:
    genome.append(line[:-1])
```

```python
gnm = ''.join(genome)
```

```python
def mirror(genome):
    mirror = {"A":"T", "C":"G", "T":"A", "G":"C"}
    mirrored_genome=[]
    for letter in genome:
        mirrored_genome.append(mirror[letter])
    mirrored_genome = ''.join(mirrored_genome)
    return mirrored_genome
```

```python
new_gnm = mirror(gnm)
```

```python
def toARN(ADN):
    ARN = []
    dna_to_rna = {"T":"U"}
    for letter in ADN:
        if letter in dna_to_rna:
            ARN.append(dna_to_rna[letter])
        else:
            ARN.append(letter)
    ARN = ''.join(ARN)
    return ARN
```

```python
arn = toARN(gnm)
```
